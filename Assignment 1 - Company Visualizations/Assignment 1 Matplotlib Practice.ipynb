```python
#Read the companies file into Pandas data frame and print a list of companies based in the USA
import pandas as pd

companies = pd.read_csv('companies.csv')

list(companies.columns)
```

```python
print(companies.head())
```

```python
companies_USA = companies[companies['country'] == 'USA'] #Selecting only companies in USA

print(companies_USA['companyName'])
```

```python
#Sort (in Python) the companies in descending order of revenue and generate a bar chart 

companies_sorted = companies_USA.sort_values(by=['annualRevenue'], ascending=False)

print(companies_sorted)


import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,6))
# make bar plot with matplotlib
plt.bar('companyName','annualRevenue',data=companies_sorted)
plt.xlabel("companyName", size=15)
plt.ylabel("Annue Revenue", size=15)
plt.title("Revenue across American Companies", size=18)
```

```python
#Generate and briefly interpret a bubble chart showing all companies by the number of employees

import seaborn as sns

plt.figure(2, figsize=(12,8))
sns.scatterplot(data=companies, 
                x="companyName", 
                y="employees", 
                sizes=(20,500),
                size="employees")

plt.xticks(rotation=90) #makes these vertical

# adding legend for the figure
plt.xlabel("Companies", size = 16)
plt.ylabel("Revenue in Billions", size = 16)
plt.title("Company Revenue according to Revenue", size = 17)


```

```python
# Append a new column that concatenates "city", "state", and "country" fields
companies['Address'] = companies['city'] + '-' + companies['state'] + '-' + companies['country']
print(companies)
```

```python
#Generate a scatterplot of revenue and the number of employees with a regression line, with meaningful chart name and axis labels. And, provide brief interpretation. 
# use the scatterplot function to build the bubble map
import numpy as np


plt.figure(2, figsize=(12,8))
x = companies['employees']
y = companies['annualRevenue']
plt.plot(x, y, 'o')

m, b = np.polyfit(x, y, 1)
plt.xlabel("Number of Employees",size = 16)
plt.ylabel("Revenue in Billion Dollars",size = 16)
plt.title("Number of Employees vs Revenue",size = 17)
plt.plot(x, m*x + b)
```

```python
#Generate two pie charts of companies by revenue and by the number of employees
#Pie Chart #1: Revenue 
companies.plot.pie(x = "companyName", y = "annualRevenue", labels = companies["companyName"], figsize = (6,6))
plt.title("Company revenue", size=18)
```

```python
#Pie Chart #2: Number of Employees

companies.plot.pie(x = "companyName", y = "employees", labels = companies["companyName"], figsize = (6,6))
plt.title("Company Employees", size=18)
```
